## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_mouse_df = pd.merge(mouse_metadata, study_results, how="inner", on="Mouse ID")
combined_mouse_df

# Display the data table for preview
```

```python
# Checking the number of mice.
mouse_count = combined_mouse_df["Mouse ID"].count()
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_rows = combined_mouse_df[combined_mouse_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_rows
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
all_duplicate_rows = combined_mouse_df[combined_mouse_df.duplicated(['Mouse ID',])]
all_duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = combined_mouse_df.drop_duplicates("Mouse ID")
clean_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = clean_mouse_df["Mouse ID"].count()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
mean = combined_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = combined_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = combined_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dv = combined_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = combined_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_mouse_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": standard_dv, "SEM": sem})
summary_mouse_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
```

```python
all_duplicate_rows.head()
```

```python
# Sort data by Drug Regime, Mouse ID and Timepoint
sorted_df = all_duplicate_rows.sort_values(["Drug Regimen", "Mouse ID", "Timepoint"], ascending=True)
# Select final volume of each mouse 
max_df = sorted_df.loc[sorted_df["Timepoint"] == 45]
max_df.head().reset_index()
```

```python
# Select data for Capomulin regimen and reset index
cap_data_df = max_df[max_df['Drug Regimen'].isin(['Capomulin'])]
cap_data_df.head().reset_index()
```

```python
# Convert column "Tumor Volume" of the Capomulin regimen into a dataframe object
cap_list = cap_data_df.sort_values(["Tumor Volume (mm3)"], ascending=True).reset_index()
cap_list = cap_list["Tumor Volume (mm3)"]
cap_list
```

```python
# Use pandas to give quartile calculations
quartiles = cap_list.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of temperatures is: {lowerq}")
print(f"The upper quartile of temperatures is: {upperq}")
print(f"The interquartile range of temperatures is: {iqr}")
print(f"The the median of temperatures is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title("Final Tumor Volume in Capomulin Regimen")
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(cap_list)
plt.show()
```
